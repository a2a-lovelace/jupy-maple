## Задание 1: Построить график функции $\sin^2(2x)$ на отрезке $[0, π]$

```python
# Подключим библиотеки
from sympy import symbols, sin, pi
from sympy.plotting import plot

# Символьная переменная
x = symbols('x')

# Наша функция
function = sin(2*x)**2

# От 0 до Пи график:
plot(function, (x, 0, pi), title='sin(2*x)**2', xlabel='x', ylabel='sin(2*x)**2', line_color='green')
```

## Задание 2. Построить график неявной функции $x^4 + y^2 = e^x$ на интервале x ∈ (−2, 2).

```python
# Подключаем библиотеку sympy
from sympy import symbols, exp, Eq
from sympy.plotting import plot_implicit

# Вводим символьные переменные
x, y = symbols('x y')

# Задаем функцию
equation = Eq(x**4 + y**2, exp(x))

# Рисуем график -2 до 2
plot_implicit(equation, (x, -2, 2), (y, -5, 5), line_color='blue')
```
